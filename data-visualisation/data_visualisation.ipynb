## MatPlotLib vs Seaborn:
[⇡ Back to top](#Python-Data-Visualisation)

* **Matplotlib is a general-purpose plotting library for Python**. 
    * *Use it for*: 
        * basic plots
        * default styling
        * graphical visualisation for numpy and pandas
        * accessing extensive library of functions, but with complex syntax
    * *Cannot use for*: 
        * plotting categorical variables with boxplots or violin plot
        * visualising regression models
        * custom styling
    
* **Seaborn is a Python data visualization library based on matplotlib**. 
    * *Use it for*: 
        * more advanced statistical plots
        * visualising regression models
        * custom styling
        * functions to plot categorical variables with boxplots and violin plots
        * compatibility with dataframes in pandas
        * simpler syntax

## <font color='blue'>2. Create a Line Chart</font>
Displays a serials of data points or markers, connected by straight lines

### Create a Line Chart using MatPlotLib:
[⇡ Back to top](#Python-Data-Visualisation)

* Create a **list** showing yield of apples (tons per hectare) over 6 years in Kristyland

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

* visualise how the yield of apples change over time using a line chart using **matplotlib** (plt)

```python
plt.plot(yield_apples)
```

* just output the graph (**remove the output line**):

```python
plt.plot(yield_apples);
```

* **display the year** for the x-axis (horizontal), instead of the 0-5 indicies

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
plt.plot(years, yield_apples);
```

* **add axis labels** (y and x)

```python
plt.xlabel('Year') #horizontal
```

```python
plt.ylabel('Yeild (tons per hectare)'); #vertical
```

* **combine the labels and plotted data** (in the same output)

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yeild (tons per hectare)');
```

* **plot multiple lines** of data

```python
years = range(2000, 2012)
# range shortcut - instead of years list

apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
# first data set

oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
# second data set

plt.plot(years, apples) # first data set line
plt.plot(years, oranges) # second data set line

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

* **add title and legend** to visualisaton

```python
# plot the data
plt.plot(years, apples)
plt.plot(years, oranges)

# label the axis
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

# define the visualisation
plt.title("Crop Yields in Kristyland")
plt.legend(['Apples', 'Oranges']);
```

* **add markers** to plotted data
* [marker list](https://jovian.com/outlink?url=https%3A%2F%2Fmatplotlib.org%2F3.1.1%2Fapi%2Fmarkers_api.html)

```python
# plot and mark data points
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='*')

# label the axis
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

# define the visualisation
plt.title("Crop Yields in KristyLand")
plt.legend(['Apples', 'Oranges']);
```

* **style lines and markers**
* `plt.plot`arguments:
    * `linestyle or ls` – Choose between a solid or dashed line
    * `linewidth or lw` – Set the width of a line
    * `markersize or ms` – Set the size of markers
    * `markeredgecolor or mec` – Set the edge color for markers
    * `markeredgewidth or mew` – Set the edge width for markers
    * `markerfacecolor or mfc` – Set the fill color for markers
    * `alpha` – Opacity of the plot
    * `color or c` – Set the color of the line ([supported colors](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
    
![image](matplotlib_colors.png)

```python
# plot mark and style data
plt.plot(years, apples, marker='*', c='lawngreen', ls='-', lw=2, ms=15, mew=1, mec='navy')

# ls = line style solid
# lw = line width of 2px
# ms = marker size of 15px
# mew = marker edge (outline) width of 1px
# mec = marker edge (outline) color navy

plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

# ls = line style dashed
# lw = line width of 3px
# ms = marker size of 10px
# alpha = marker plot opacity 50%

# label the axis
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

# define the visualisation
plt.title("Crop Yields in KristyLand")
plt.legend(['Apples', 'Oranges']);
```

* **specify marker shape, line style and line color** in shorthand

```python
fmt = '[marker][line][color]' # shorthand to reference in plt.plot argument

plt.plot(years, apples, '*-b') # s = star marker, - = solid line style , b = blue line color
plt.plot(years, oranges, 'o--r') # o = circle marker, - = dashed line style , r = red line color

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in KristyLand")
plt.legend(['Apples', 'Oranges']);
```

* **change figure size** in visualisation

```python
plt.figure(figsize=(12, 6)) # x, y axis

plt.plot(years, oranges, 'or') # o = circle marker, r = color marker red, no line to connect
plt.title("Yield of Oranges (tons per hectare)");
```

### Improve the Line Chart using Seaborn:
[⇡ Back to top](#Python-Data-Visualisation)

* review the [pre-defined styles](https://seaborn.pydata.org/generated/seaborn.set_style.html) set by Seaborn
* **incorporate a grid style** into your line chart

```python
sns.set_style("whitegrid") # light background
plt.plot(years, apples, 's-b') # marker = square, line = solid, color = blue
plt.plot(years, oranges, 'o--r') # marker = circle, line = dashed, color = red

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kristyland")
plt.legend(['Apples', 'Oranges']);
```

```python
sns.set_style("darkgrid") # dark background
plt.plot(years, apples, 's-b') # marker = square, line = solid, color = blue
plt.plot(years, oranges, 'o--r') # marker = circle, line = dashed, color = red

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kristyland")
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, oranges, 'or') # marker = circle, color = red
plt.title("Yield of Oranges (tons per hectare)");

# note: continues last sns.set_style action run
```

* **edit default style** by modifying the 'matplotlib.rcParams' dictionary. [Read More...](https://matplotlib.org/3.2.1/tutorials/introductory/customizing.html#matplotlib-rcparams)

```python
import matplotlib as mpl

mpl.rcParams['font.size'] = 20
mpl.rcParams['figure.figsize'] = (10, 10)
mpl.rcParams['figure.facecolor'] = '#00000000'
mpl.rcParams['lines.linestyle'] = '-'
mpl.rcParams['lines.linewidth'] = 5
```

```python
plt.plot(years, oranges)
plt.title("Yield of Oranges (tons per hectare)");
```

```python
mpl.rcParams['font.size'] = 14
mpl.rcParams['figure.figsize'] = (9, 5)
mpl.rcParams['figure.facecolor'] = '#00000000'
mpl.rcParams['lines.linestyle'] = '--'
mpl.rcParams['lines.linewidth'] = 2
```

```python
plt.plot(years, oranges)
plt.title("Yield of Oranges (tons per hectare)");
```

## <font color='blue'>3. Create a Scatter Plot</font>

### Scatter Plots in MatPlotLib:
[⇡ Back to top](#Python-Data-Visualisation)

* the values of 2 variables are plotted as points
    * these are on a 2-dimensional grid
* use a 3rd variable to style points

* Use the [iris flower dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set) (included in Seaborn library) to **load data as Pandas dataframe**

```python
# load data into Pandas dataframe:
flowers_df = sns.load_dataset("iris")

# seaborn.load_dataset = function that returns a pandas dataframe object
```

```python
flowers_df
```

* **print the unique values from an array** based on a specified field

```python
flowers_df.species.unique()
```

* **visualise the relationship between 2 values:** sepal length and width

```python
plt.plot(
    flowers_df.sepal_length,
    flowers_df.sepal_width
);
```

Note: there isn't a simple relationship -> choose a different visualisation for clarity

* **visualise the data using a scatterplot** with seaborn (sns)

```python
sns.scatterplot(
    x=flowers_df.sepal_length,
    y=flowers_df.sepal_width,
);
```

### Add Hues in MatPlotLib:
[⇡ Back to top](#Python-Data-Visualisation)

* Points above form clusters with some outliers
* Color the dots to **identify the clusters**

```python
sns.scatterplot(
    x=flowers_df.sepal_length,
    y=flowers_df.sepal_width,
    hue=flowers_df.species, #add the hue based on specified category
    s=100 #s = size of the markers... eg 100px
);
```

### Customize Seaborn Figures:
[⇡ Back to top](#Python-Data-Visualisation)

```python
# use pyplot to customise figure:

plt.figure(figsize=(12, 6)) 
#change figure size (note: default was changed prior with: mpl.rcParams['figure.figsize'] = (9, 5))

plt.title('Sepal Dimensions')

# use seaborn to create scatterplot:
sns.scatterplot(x=flowers_df.sepal_length, 
                y=flowers_df.sepal_width, 
                hue=flowers_df.species,
                s=100);
```

### Plot Data using Pandas Data Frames with Seaborn:
[⇡ Back to top](#Python-Data-Visualisation)
* Seaborn as built in support for Pandas dataframes
    * no need to pass each column as a series
        * provide column names and use data argument to specify data frame

```python
# set scatterplot title with pyplot:
plt.title('Sepal Dimensions')

# set scatterplot dimensions, hue of individual dot values based on column, 
    # size of markers and data to specify dataframe 
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species', # hue of dot based on unique values in species column
                s=100, # size of marker dot
                data=flowers_df); #data to access for markers
```
