## Aggregating Data with Python

```python
import pandas as pd
```

```python
# create dataframe
df = pd.read_csv("MonthlyProductSales.csv", encoding = "ISO-8859-1")
```

```python
# print the top row
df.head(n=1)
# note: the date format
```

```python
# look @ column month of order date - only display the first 4 characters (the year)
df['Month of Order Date'].str[:4]
```

```python
# Group by the year, and provide descriptive stats (similar to SQL)
df.groupby(df['Month of Order Date'].str[:4]).describe()
```

```python
df
```

```python
# reset the index values of the df
df.groupby(df['Month of Order Date'].str[:4]).describe().reset_index()
```

```python
df
```

```python
# rename the month of order day column to year of order date
df.groupby(df['Month of Order Date'].str[:4]).describe().reset_index().rename(columns={"Month of Order Date":"Year"})
```

```python
df
```

```python
# get sum of product sales by year
# group df by Month of order date, display only the year and product name column. 
# sum the values for each product in each year
# reset index for this df

yearly_product_sales = df.groupby([df["Month of Order Date"].str[:4], "Product Name"]).sum().reset_index()
# doesn't work due to duplicate name
```

```python
yearly_product_sales = df.groupby([df["Month of Order Date"].str[:4].rename("Year"), "Product Name"]).sum().reset_index()
```

```python
yearly_product_sales.rename(columns={"Month of Order Date":"Year"})
```

```python
df.groupby("Product Name").sum().reset_index()
```

```python
yearly_product_sales = df.groupby([df["Month of Order Date"].str[:4].rename("Year"), "Product Name"])["Sales"].sum().reset_index()
```

```python
yearly_product_sales
```

```python
yearly_product_sales.to_csv("yearly_product_sales.csv", header=True, index=False, encoding="utf-8")
```

## Visualising Data

### Sales Analysis Data
Analysing monthly product sales in 3 ways:
- plot
- histogram
- horizontal bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import CSV of monthly sales
df = pd.read_csv("MonthlyProductSales.csv", encoding = "ISO-8859-1")
```

```python
df.groupby(df['Month of Order Date']).sum().plot()
```

```python
# HISTOGRAM
df.groupby(df['Month of Order Date']).sum().plot(kind='hist', bins=10) # sum of sales by month
```

```python
df.groupby(df['Product Name']).sum().sort_values(by='Sales').tail(10).plot(kind='barh')
# horizontal bar chart, group by product and sales. Sort in order of sales. Top 10 Sales by product
```
